```python
import spacepy.pycdf as cdf
import numpy as np
import matplotlib.pyplot as plt

# Open the CDF file
cdf_file = cdf.CDF('26006/solo_L1_mag-obs-normal_20230626_V01.cdf')
```

```python
# Read the data
time = cdf_file['EPOCH'][:]
B_OBS1 = cdf_file['B_OBS_URF'][:,0]
B_OBS2 = cdf_file['B_OBS_URF'][:,1]
B_OBS3 = cdf_file['B_OBS_URF'][:,2]
```

```python
import matplotlib.pyplot as plt
import h5py

# Open the HDF5 file
hdf5_file = h5py.File('profiles/profile_0.h5', 'r')

# Read the data
time_1 = hdf5_file['time'][:]
B_OBS1_1 = hdf5_file['B_OBS1']
B_OBS2_1 = hdf5_file['B_OBS2']
B_OBS3_1 = hdf5_file['B_OBS3'][:]

# Plot the data
time_1_numeric = np.array([dt.timestamp() for dt in time_1])
```

```python

plt.plot(time_1_numeric, B_OBS3_1, label='B_OBS3')
plt.xticks(np.arange(time_1_numeric[0], time_1_numeric[-1], 60), rotation=45)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import h5py

# Open the HDF5 file
hdf5_file = h5py.File('profiles/profile_10.h5', 'r')

# Read the data
time_1 = hdf5_file['time'][:]
B_OBS1_1 = hdf5_file['B_OBS1']
B_OBS2_1 = hdf5_file['B_OBS2']
B_OBS3_1 = hdf5_file['B_OBS3'][:]

import matplotlib.dates as mdates
time_numeric = mdates.date2num(time_1)

plt.plot(time_numeric, B_OBS3_1, label='B_OBS3')
plt.legend()
plt.show()
```
